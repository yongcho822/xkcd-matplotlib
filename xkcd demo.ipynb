###An XKCD-Matplotlib Demo!

```python
#typical import fun
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#the magic line - enables some rcParams settings
plt.xkcd()


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#just making sure the axis only show up on the left and bottom
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#setting ticks and ticklabels blank
plt.xticks([])
plt.yticks([])

ax.set_ylim([-30, 10])

#making a data array of just a  hundred 1s.
data = np.ones(100)

#after the 70th element, we want to subtract numbers 0 to 29 from our array of 1s, one at a time.
data[70:] -= np.arange(30)

#plopping in some text at (15,-10) and drawing an arrow to (70,1)
plt.annotate(
    'the day i realized\ni could cook bacon\nwhenever i wanted',
    xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

plt.xlabel('time')
plt.ylabel('my overall health')

plt.plot(data)
```

```python
#Could xkcd make even a *gasp* pie chart bearable?

#bring it back, bring it back
plt.xkcd()

labels = '10/10\nwould', 'wow\nmuch awesome\nvery skill', 'lulz no', 'matplotlib\'s already\ntoo wild for me'
sizes = [45, 30, 15, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0.1, 0, 0, 0) # only "explode" the first slice (i.e. '10/10 would')

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')

plt.title("How we're feeling about living\nthe XKCD Matplotlib life:", y=1.1)

plt.show()
```
